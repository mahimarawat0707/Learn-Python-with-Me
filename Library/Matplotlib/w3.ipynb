```python
ypoints = np.array([2,5,7,8,10])
plt.title("main graph")
plt.plot(ypoints, 'P:y', ms=10, mec='r', mfc= 'k') #'-'	Solid line	':'	Dotted line	'--'	Dashed line	'-.'	Dashed/dotted line
plt.xlabel("Number of students present")
plt.show() #ms = markdown size, mec= marker edge color, mfc= marker face color
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1,  loc = 'left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.grid(axis= 'x', color = 'green', linestyle = '--', linewidth = 0.5)
plt.grid(axis= 'y', color = 'red', linestyle = '-.', linewidth = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["m", "hotpink", "r", "#AFAD4C"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.legend(title = "Four Fruits:")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```
