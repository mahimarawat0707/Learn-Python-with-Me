```python
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

The way of styling themes are as follows: 
white, 
dark,
whitegrid, 
darkgrid, 
ticks 

set_context() allows us to override default parameters. This affects things like the size of the labels, lines, and otherr elements of the plot, but not the overall style -> set_style(). The context are poster.   
Paper - Ideal for printing in academic papers.  
Notebook - Good for working inside Jupyter notebooks (default setting).  
Talk - Good for presentation slides where readablity matters.  

# different context

```python
sns.set_style("ticks")
sns.set_context("talk")
sns.scatterplot(data=penguins, x="flipper_length_mm", y= "body_mass_g", hue = "island")
```

```python
sns.set_style("ticks")
sns.set_context("paper")
sns.scatterplot(data=penguins, x="flipper_length_mm", y= "body_mass_g", hue = "island")
```

```python
sns.set_style("ticks")
sns.set_context("poster")
sns.scatterplot(data=penguins, x="flipper_length_mm", y= "body_mass_g", hue = "island")
```

```python
sns.set_style("ticks")
sns.set_context("notebook")
sns.scatterplot(data=penguins, x="flipper_length_mm", y= "body_mass_g", hue = "island")
```
